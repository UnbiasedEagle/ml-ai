## Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
plt.plot([1,2,3,4,5])
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
```

```python
plt.plot(x,y);
```

```python
# 1st method
fig=plt.figure() #creating a figure
ax=fig.add_subplot() #add some axes
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
#3rd method
fig,ax=plt.subplots()
ax.plot(x,y);
```

## Matplotlib Workflow Example

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
x=[1,2,3,4]
y=[11,12,13,14]
ax.set(title="Sample Plot",xlabel="x axis",ylabel="y axis")
ax.plot(x,y)
fig.savefig("./sample-plot.png")
```
